# A Project From Maven Analytics: The Maven Pizza Challenge 

### Facts
   In this challenge, I am playing the role of a BI Consultant hired by Plato's Pizza, a Greek-inspired pizza place in New Jersey. I have been hired to help the restaurant use data to improve operation.
   
_The Dataset_
       
    1)Contains 4 tables in CSV format
    2)The Orders table contains the date & time that all table orders were placed
    3)The Order Details table contains the different pizzas served with each order in the Orders table, and their quantities
    4)The Pizzas table contains the size and price for each distinct pizza in the Order Details table, as well as its
        broader pizza type
    5)The Pizza Types table contains details on the pizza types in the Pizzas table, including their name as it appears on
        the menu, the category it falls under, and its list of ingredients

### Problem Statement

Find opportunities to drive more sales, and improve work efficiency.
Some of the questions that need to be answered are:
    
    1) What days and times we tend to be the busiest?
    2) How many pizzas are we making during peak periods?
    3) What are the best and worst selling pizzas?
    4) What's our average order value?
    5) How well are we utilizing our seating capacity?

##### Primary Objective:
    
    Analyze the data and put together to help find opportunities to drive more sales and improve work efficiency.

##### Secondary Objectives:
    
    1) Determine the busiest days and times;
    2) Determine number of pizzas being made during peak periods;
    3) Identify the best and worst selling pizzas;
    4) Calculate average order value;
    5) Assess seating capacity utilization (the restaurant has 15 tables and 60 seats); and,
    6) Create a single-page visual

### ANALYSIS

#### I. BUSIEST DAYS AND TIMES

 - Friday was the busiest day of the week, with 70 daily average no. of orders, while Sunday was consistently the least.

 - Zooming into intraday, 12PM and 1PM had the most traffic throughtout the year,followed by 5PM and 6PM, while 9AM, 10AM and 11PM had the least.

##### Recommendation:

 - Concentrate manpower around 12PM to 1PM, and 5PM to 6PM everyday, but lessen number of personnel on Sundays.

#### II. PIZZAS MADE DURING PEAK PERIODS

 - Month of July sold 4,392 pizzas, the highest throughout the year, while September and October had the least, with only 3,890 and 3,883 pizzas sold respectively.
 - Friday, which was the busiest of the week, had 164 average number of pizzas sold, followed by Satruday with 144, and Thursdat having 143.  However, Sunday only had 116 average no. of pizzas sold, which was almost 50 pizzas less than Friday's.
 - On a daily basis, 12PM and 1PM were always the busisest, with an average no. of pizzas sold of 18 and 17, respectively.  These are followed by 5PM and 6PM, with 14 and 15 average no. of pizzas sold.
 - 9AM only hadhaving  1 transaction, 10AM had 8 days of transaction, and finally, 11PM with 26 days of transaction all throughout the year, were the most unproductive times of every day.
 
##### Recommendation:
 - Consider dropping off 9AM, 10AM and 11PM from operation, and only open between 11AM to 10PM.

#### III. BEST AND WORST SELLING PIZZAS

- Classic Deluxe and Barbecue Chicken Pizzas were the most popular in terms of orders and sales, but were toppled by Thai Chicken Pizza when it comes to revenue.
 - Brie Carre Pizza was consistently the least popular in all 3 areas.
 - Customers also seemed to prefer the Classic Pizzas category, having a total of 14,888 pizzas sold and the highest revenue of $220,053 generated for the whole year.

 - The Veggie Category, although it only generated $193,690 in revenue, the least among the four (4) categories, was actually slightly more popular in terms of number of orders and pizzas sold compared to the Chicken Category.
 - Large, Medium, and Small sizes were the most preferred.  Large sized pizzas were the most ordered, having a total of 18,526 orders throughout the year, followed by medium and small, with 15,385 and 14,137 orders respectively.  XL and XXL, however, only got 544 and 28 orders respectively, for the whole year.

##### Recommendations:
 - Consider offering Specialty Pizzas with a large variety /combination of toppings from popular pizzas.
 - Price increase on popular pizzas may be implemented gradually and selectively, while closely monitoring customers' behavior.
 - Dropping off Brie Carre Pizza from the menu may improve inventory cost and management.
 - XL and XXL sizes of popular pizzas may also be introduced.

#### IV. ABOVE AVERAGE VALUE
 - Total Revenue =  $817,860
 - Total Orders = 21,350
 
 - Average Order Value = $38

#### V. SEATING CAPACITY UTILIZATION


 - There were only 17 instances in hourly timeframe when the number of customers exceeded the restaurant's seating capacity.  And, this occured  on 12PM and 1PM timeframe.
 
##### Recommendations:
 - Consider recording "take-aways" separately from "dine-ins".  This will significantly improve anlysis on seating capacity utilization.
 - Since most ordered pizzas were Small to Large sizes, we can also assume that these are for 1 to 2 persons only.  Providing 2 seater tables will also maximize the seating capacity of the restaurant.

### The Seating Capacity

The restaurant has a total of 15 tables and 45 chairs.

Aside from this info, there are no other available data that is directly relevant in determining the utilization of the restaurants seating capacity.

With this, I have to incorporate into my analyses the following assumptions:

    1. All orders are "for dine-ins"
    2. The pizza to person ratio:
        Small  = 1 person in average
        Medium = 1 person in average
        Large  = 2 persons in average
        XL     = 3 persons in average
        XXL    = 4 persons in average

```python
size = pd.DataFrame(pizzas.groupby(by = "size"))
size['Pax'] = ['2', '1', '1','3','4']
size
```

```python
size.columns.values
```

```python
size = size.drop([1], axis =1)
```

```python
size = size.rename(columns = {0: 'Size'})
```

```python
size['Pax'] = size['Pax'].astype(int)
size
```

```python
pizza_sales_new = pd.merge(pizza_sales,size)
pizza_sales_new.sort_values(by = 'Pizzas Sold', ascending=False)
```

```python
pizza_sales_new['Total Pax'] = pizza_sales_new['Pizzas Sold'] * pizza_sales_new['Pax']
pizza_sales_new
```

```python
pizza_sales_new.describe()
```

```python
pax = pizza_sales_new.groupby(by = ['Date','Day Name', 'Day Time'], 
                              as_index = False).agg({'Pizzas Sold':'sum', 
                                                     'Amount': 'sum', 
                                                     'Total Pax': 'sum'})

pax = pax.astype({'Pizzas Sold': 'int', 'Amount' : 'int', 'Total Pax': 'int'})
pax.head()
```
